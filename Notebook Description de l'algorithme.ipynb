## Présentation de la mission

De nos jours, nous ne pouvons ignorer ces différents enjeux : la consommation d’énergie, l’économie et le recyclage des matières premières, l’amélioration des modes de transports...

Notre ambition est de bouleverser le marché en faisant évoluer les comportements pour réduire la consommation d’énergie et des émissions de gaz à effet de serre.

Pour cela, il faut que nous optimisions nos tournées : ça tombe bien, nous ne sommes pas les seuls à nous être penché sur le problème.

C'est en fait le **problème de tournées de véhicules (VRP)**, qui a pour but de minimiser le coût de livraison des biens. 

Notre problème algorithmique est donc de calculer sur un réseau routier une tournée permettant de relier entres elles nos clients, puis de revenir à notre dépôt, de manière à minimiser la distance totale parcourue. 

Le problème de tournées de véhicules est un **problème d'optimisation combinatoire** et de **recherche opérationnelle**.

Il existe des variantes à ce problème VRP, et nous nous sommes penchés sur la variante suivante :

- **Problème de routage de véhicule par capacité: CVRP.** Cette variante ajoute une **contrainte supplémentaire** : les véhicules ont une capacité de charge limitée des marchandises à livrer.

## Modélisation du problème algorithmique

Pour nos jeunes entrepreneurs, les choses sérieuses commencent.

Il faut maintenant modéliser le problème sous une forme mathématique.

En faisant leurs recherches, ils apprennent que tout problème d'optimisation combinatoire se doit d'être modélisé en 4 étapes :
- Définition du problème formel
- Variable de décision
- Fonction objective
- Etablissement des contraintes

### Définition du problème formel

Un dépôt a un ensemble de clients qui lui est associé.

Le problème de tournées est défini sur un graphe $$G = {V,A}$$
où

$$ V = \left \{v_0,v_1,...,v_n\right \} $$ est l'ensemble des $n$ clients que nous devons livrer. Ils sont représentés par les sommets sur notre graphe.

$$ A = \left \{(v_i,v_j):v_i,v_j\right \} $$
avec $v_i,v_j \in V$ et $i \neq j$, est l'ensemble des trajets en camion entre les différents clients où chaque trajet $(v_i,v_j)$ est associé à un coût de transport $(c_{ij})$.

Chaque dépôt possède un ensemble limité de véhicules $(k=1,..,K)$.
K est le nombre maximal de véhicules disponible.

Dans une route r du véhicule k, un client ne peut être visité qu'une et une seule fois.

### Les paramètres

- Le nombre de clients : $n$ (Ensemble $N$)
- Le nombre de camions : $k$ (Ensemble $K$)
- La capacité des véhicules : $Q$
- La demande du client : $q_i$
- La longueur maximale d'une route : $D$
- Le coût de l'arrête entre deux sommets $i$ et $j$ (distance de parcours : $c_{ij}$

précision : $\sum$ est la lettre sigma majuscule symbolisant l'addition de termes consécutifs. En bas du signe se trouve le premier indice de la série. En haut du signe, c'est le dernier indice de la série.

### Variable de décision

Pour chaque trajet entre deux clients, on cherche à savoir si une route r va choisir ce trajet.
Ce choix est représenté par la variable de décision.

$$x_{ijk}=\left\{\begin{matrix}
1 \text{ si } (i,j) \text{ est parcouru par le vehicule } k
\\0 \text{ sinon}
\end{matrix}\right.
$$

### Fonction objective

La fonction objective sert à définir l'objectif de l'algorithme.

Nous cherchons à minimiser la somme des coûts pour la tournée de livraison des clients.

$$min \sum_{i=0}^{n} \sum_{j=1}^{n+1} c_{ij} \sum_{k \in K} x_{ijk}$$

### Contraintes

1.
$$\sum_{j=1,j \neq i}^{n} \sum_{k \in K} x_{ijk} = 1$$
2.
$$\sum_{i=1,i \neq j}^{n} \sum_{k \in K} x_{ijk} = 1$$

Sur ces deux contraintes, on indique que **chaque client est desservi une et une seule fois** par **un et un seul camion**.

La contrainte 1 assure qu'**un seul trajet quitte un client i vers un quelconque autre client j**. Il assure donc la **continuité logique du trajet.**

La contrainte 2 assure qu'**un seul trajet relie un client j afin de répondre à la demande du client.**

Ceci assure en même temps l'équilibre. Il y a **conservation des flux**.

3.
$$\sum_{j=1}^{n} x_{ojk} = 1$$
4.
$$\sum_{i=1}^{n} x_{i0k} = 1$$

Les contraintes 3 et 4 assurent que **chaque trajet commence et termine au dépôt.**

5.
$$x_{ijk}=\left \{ 0,1 \right \}$$

La contrainte 5 affirme que **les variables d'arcs sont des variables binaires.**

6.
(CVRP)
$$\sum_{i=1}^{n} \sum_{j=1}^{n} q_{i} x_{ijk} \leq Q$$

La contrainte 6 assure que **l'ensemble des livraisons effectués aux clients visités par le véhicule respecte la capacité de ce dernier.**

7.
(VRPTW)
$$\sum_{i=1}^{n} \sum_{j=1}^{n} c_{ij} x_{ijk} \leq D$$

La contrainte 7 assure que le **respect du temps limite qu'un trajet peut prendre.**


### Etude de la complexité

Le problème de tournée de véhicules (VRP) est un problème combinatoire impossible à résoudre optimalement dans un temps polynomial. Il est toutefois possible de vérifier une solution à ce problème en temps polynomial. On le classe ainsi en tant que _problème NP_.

De plus, tous les problèmes NP tels que le problème du voyageur de commerce se réduisent via une réduction polynomial à notre problème VRP. Nous le classifions alors comme un **problème de décision NP-Complet.**

Pour expliquer et modéliser plus simplement la difficulté de ce problème, nous pouvons prendre le VRP sous sa forme la plus simple où il n’y a qu’un seul véhicule qui dessert tous les clients, équivalent au problème du voyageur de commerce. 

Pour un problème de voyageur de commerce, le nombre de solutions différentes possible est donné par 

$$½ n!$$

- n est le nombre de nœuds à visiter
- n! est le nombre de routes différentes possibles
- le facteur ½ vient du fait que les graphes ne sont par orienté, autrement dit, le sens de circulations n’a pas d’importance.

Les algorithmes connus pour résoudre le VRP ont une complexité temporelle exponentielle. Ceci signifie qu’il existe des instances pour lesquelles l’algorithme prendra un **temps exponentiel.**

Vu que la complexité d’un problème est la complexité minimale des algorithmes permettant de résoudre le problème, on dira ici que notre problème de tournée de véhicules possède une **complexité exponentielle.**

Pour résoudre notre problème de tournée de véhicules (VRP), il existe différentes méthodes de résolution. 

- Il est possible d’utiliser des **méthodes approchées à base d’heuristique** pour obtenir des solutions dans un temps de calcul raisonnable. 

- Il y a également la possibilité d’utiliser des **métaheuristiques et des algorithmes génétiques** ou de réaliser de **l’optimisation linéaire** qui permet de résoudre de façons **exacte** certains problèmes de tournées de véhicules de taille réduite.

### Description de l'algorithme

D’abord on implante le package **numpy** qui nous servira à générer des nombres aléatoires. 

Puis on initialise nos variables comme suit :
- n : nombre de clients
- Q : Capacité maximum du camion
- N : Tous les nœuds.
- V : Tous les nœuds avec le dépôt.
- d : une collection A qui contient la demande de chaque client

On initialise ensuite un emplacement aléatoire pour chaque nœud et le dépôt sur un graphique de dimension aléatoire puis on l’affiche.

Puis on initialise les arrêtes entre chaque nœud avec pour chaque arrêtes, leur coût c.

On utilise ensuite **l’API CPLEX** pour initialiser notre fonction objective et nos contraintes et afficher la solution. L'algorithme implémenté ici est celui du **Branch & Cut.**

Dans la solution, nous pouvons voir le coût total (distance totale parcourue), et également voir quels clients sont reliés. Nous voyons également la demande cumulée pour chaque client, selon la route sur laquelle il se trouve.

À la suite de ça, on réaffiche le graphique avec les arrêtes actives, c’est-à-dire celle qui sont utilisées pour résoudre le problème. Cela nous permet d’avoir un visuel sur la solution.
